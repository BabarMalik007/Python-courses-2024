# Encapsulation

```python
class Laptop():
    def __init__(self,comp, model, gen, storage, ram,pro,price):
        self.comp = comp
        self.model =model
        self.gen = gen
        self.storage = storage
        self.ram =ram
        self.pro =pro
        self.price = price
    def specs(self):
        print(f"""
        
            Company : {self.comp}
            Model:    {self.model}
            Generation: {self.gen}
            Storage:    {self.storage}
            Ram :       {self.ram}
            Processor:   {self.pro}
            Price:     {self.price}""")
    def play_games(self):
        print("I am playing games")
           
    def get_comp(self):
        return self.comp
    def set_comp(self,new_comp):
        companies = ['lenovo', 'dell', 'hp']
        if new_comp in companies:
            self.comp = new_comp
        else:
            print("Not a valid company")
```

### Modifying an Attributeâ€™s Value Through a Method

```python
class Car():
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading=0        
    
    def get_descriptive_name(self):
        return f"{self.year} {self.make} {self.model}".title()
    def set_odometer(self, new_reading):
        if new_reading < self.odometer_reading:
            print("Cheater insan ban ja.")
        else:
            self.odometer_reading = new_reading

    def get_odometer(self):
        print(f"This car has {self.odometer_reading} miles on it.")
```

```python
my_new_car = Car('audi', 'a4', 2016)
print(my_new_car.get_descriptive_name())
```

```python
my_new_car.get_odometer()
```

```python
my_new_car.set_odometer(10)
```

```python
my_new_car.get_odometer()
```

### Overriding Methods from the Parent Class

```python
class Car():
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading=0        
    
    def get_descriptive_name(self):
        return f"{self.year} {self.make} {self.model}".title()
    def set_odometer(self, new_reading):
        if new_reading < self.odometer_reading:
            print("Cheater insan ban ja.")
        else:
            self.odometer_reading = new_reading

    def get_odometer(self):
        print(f"This car has {self.odometer_reading} miles on it.")
        
    def fill_gas_tank(self):
        print("OK car tank is being filled with gasoline")
```

```python
class ElectricCar(Car):
    def __init__(self, make, model, year):
        super().__init__(make, model, year)
        self.battery_size =  70
    def describe_battery(self):
        print("This car has a " + str(self.battery_size) + "-kWh battery.")
    def fill_gas_tank(self):
        print("There is no gas tank needed in Electric Car")
```

```python
my_tesla = ElectricCar('tesla', 'model s', 2016)
```

```python
my_tesla.fill_gas_tank()
```
